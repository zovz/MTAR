```python
import pandas as pd
```

```python
# Load the datasets
with open("Normal_120s.pkl", "rb") as f:
    normal_data = pd.read_pickle(f)

with open("Vibration_120s.pkl", "rb") as f:
    vibration_data = pd.read_pickle(f)

with open("RampUpDown_300s.pkl", "rb") as f:
    rampupdown_data = pd.read_pickle(f)

with open("Friction_120s.pkl", "rb") as f:
    friction_data = pd.read_pickle(f)
```

# The normal data

The normal data is 2 minutes long, containing 120 000 samples for each of the three axis for the two sensors.

The data is collected running the motor without any disturbances.

```python
normal_data
```

# Vibration Data

The vibration data is 2 minutes long, containing 120 000 samples for each of the three axis for the two sensors.
The data is retrieved by manually adding vibrations by placing a metal object on the axel of the motor, causing vibrations. 

```python
vibration_data
```

# Friction Data

The friction data is 2 minutes long, containing 120 000 samples for each of the three axis for the two sensors.

The data is collected by slightly twisting the bearing, causing friction in the bearing.

```python
friction_data
```

# Ramp up-down Data

The ramp up-down data is 5 minutes long, containing 300 000 samples for each of the three axis for the two sensors.

The data is collected by ramping up the motor to full speed, and then ramping it down to almost a stop over and over again for the duration of the dataset. 

```python
rampupdown_data
```
